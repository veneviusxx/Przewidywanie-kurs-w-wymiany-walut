```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Wczytanie danych z pliku CSV
file_path = r'C:\Users\justy\Desktop\Dane_usdeur\usdeur_przedpandemia.csv'
przedpandemia = pd.read_csv(file_path)

# Konwersja kolumny 'Date' na typ daty
przedpandemia['Date'] = pd.to_datetime(przedpandemia['Date'])

# Zmiana nazwy kolumny 'Close*' na 'Close'
przedpandemia.rename(columns={'Close*': 'Close'}, inplace=True)

# Wykresy czasowe
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=przedpandemia)
plt.title('USD/EUR - Wykres czasowy')
plt.xlabel('Data')
plt.ylabel('Kurs USD/EUR')

# Zapisz pierwszy wykres
plt.savefig('wykres_czasowy.png')
plt.close()

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(przedpandemia['Close'], kde=True)
plt.title('USD/EUR - Histogram')
plt.xlabel('Kurs USD/EUR')
plt.ylabel('Liczba obserwacji')

# Zapisz drugi wykres
plt.savefig('histogram.png')
plt.close()

# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=przedpandemia['Date'].dt.year, y=przedpandemia['Close'])
plt.title('USD/EUR - Boxploty dla poszczegÃ³lnych lat')
plt.xlabel('Rok')
plt.ylabel('Kurs USD/EUR')

# Zapisz trzeci wykres
#plt.savefig('boxploty.png')
#plt.close()
plt.show()
```

```python
# Wczytanie danych z pliku CSV
file_path_wtrakciepandemii = r'C:\Users\justy\Desktop\Dane_usdeur\usdeur_wtrakciepandemii.csv'
wtrakciepandemii = pd.read_csv(file_path_wtrakciepandemii)

# Konwersja kolumny 'Date' na typ daty
wtrakciepandemii['Date'] = pd.to_datetime(wtrakciepandemii['Date'])

# Zmiana nazwy kolumny 'Close*' na 'Close'
wtrakciepandemii.rename(columns={'Close*': 'Close'}, inplace=True)

# Wykresy czasowe
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=wtrakciepandemii, color='purple')
plt.title('USD/EUR - Wykres czasowy (W trakcie pandemii)')
plt.xlabel('Data')
plt.ylabel('Kurs USD/EUR')
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(wtrakciepandemii['Close'], kde=True, color='purple')
plt.title('USD/EUR - Histogram (W trakcie pandemii)')
plt.xlabel('Kurs USD/EUR')
plt.ylabel('Liczba obserwacji')
plt.show()

# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=wtrakciepandemii['Date'].dt.year, y=wtrakciepandemii['Close'], color='purple')
plt.title('USD/EUR - Boxploty dla poszczegÃ³lnych lat (W trakcie pandemii)')
plt.xlabel('Rok')
plt.ylabel('Kurs USD/EUR')
plt.show()
```

```python
# Wczytanie danych z pliku CSV
file_path_odwojny = r'C:\Users\justy\Desktop\Dane_usdeur\usdeur_odwojny.csv'
odwojny = pd.read_csv(file_path_odwojny)

# Konwersja kolumny 'Date' na typ daty
odwojny['Date'] = pd.to_datetime(odwojny['Date'])

# Zmiana nazwy kolumny 'Close*' na 'Close'
odwojny.rename(columns={'Close*': 'Close'}, inplace=True)

# Wykresy czasowe
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=odwojny, color='#FF8C00')
plt.title('USD/EUR - Wykres czasowy (Od wojny)')
plt.xlabel('Data')
plt.ylabel('Kurs USD/EUR')
plt.show()

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(odwojny['Close'], kde=True, color='#FF8C00')
plt.title('USD/EUR - Histogram (Od wojny)')
plt.xlabel('Kurs USD/EUR')
plt.ylabel('Liczba obserwacji')
plt.show()

# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=odwojny['Date'].dt.year, y=odwojny['Close'], color='#FF8C00')
plt.title('USD/EUR - Boxploty dla poszczegÃ³lnych lat (Od wojny)')
plt.xlabel('Rok')
plt.ylabel('Kurs USD/EUR')
plt.show()
```
